I want to see only the rows where AWATER is greater than 0. First, I define a variable, called filteredAWATER, and set this variable equal to all of the rows in the boundaries dataframe where the AWAWTER column is greater than 0. However, AWATER is a series and the fields are objects, whereas 0 is an integer. We can't compare a series/object to an integer, so we must convert AWATER to an integer. Therefore, we make a copy of AWATER with all values converted to integers. 


```python
# convert AWATER data type to integers and return all rows with value > 0
## set filteredAWATER equal to all of the rows in boundaries where the AWATER column is greater than 0
### In order to compare AWATER to integer '0', make a copy of AWATER with all values converted to ints

filteredAWATER = boundaries[boundaries.AWATER.astype(int) > 0]

filteredAWATER
```

The operation above displays a dataframe that returns all rows with value > 0
